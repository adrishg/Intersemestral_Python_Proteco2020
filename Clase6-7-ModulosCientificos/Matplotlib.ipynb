```python
#importar el modulo con un alias plt
import matplotlib.pyplot as plt
%matplotlib inline
#importe también numpy
import numpy as np
```

### Los básicos

```python
#Nuestros datos
x = [1,2,3]
y = [2,4,6]


plt.title("Mi primer gráfica en matplotlib", fontdict={'fontsize':18})
plt.xlabel("Eje x")
plt.ylabel("Eje y")

plt.plot(x,y, label="2x", color="g", linewidth=1, marker="*", markersize=10)

plt.legend()
plt.show()
```

```python
x= np.linspace(0, 10, 10)

plt.figure(figsize=(2,3), dpi=150)

plt.plot(x,x, label= "Línea recta", color="lawngreen", linestyle="-")
plt.plot(x, x**2, label= "x al cuadradp", color="hotpink", linestyle="-.", marker="^")
plt.plot(x, x**3, "b*--" , label="x al cubo")



plt.legend()
plt.show()


```

### Gráficas de barra

```python
etiquetas= ['A','B','C','D']
values= [10,7,4,8]

plt.bar(etiquetas, values, color=['black','green','hotpink','yellow'])
plt.show()
```

```python
estaturasMujeres= [150, 167, 180,166,156]
estaturasHombres= [170, 160,178,183,145]

rangoAlturas=np.linspace(140,190, num=5)

fig= plt.figure()
ax= fig.add_axes([0,0,1,1])
plt.scatter(rangoAlturas, estaturasMujeres, color="r")
plt.scatter(rangoAlturas, estaturasHombres, color="y")

plt.show()
```

```python
datos= {'cucharas':4, 'tenedores':5,'cuchillos':3, 'platos':8}
nombres= list(datos.keys())
valores= list(datos.values())

fig, ((ax1, ax2),(ax3,ax4))= plt.subplots(2,2, figsize=(8,8), sharey=False)

ax1.bar(nombres,valores)
ax1.set_title("Gráfica de barras")

ax2.scatter(nombres,valores)
ax2.set_title("Gráfica de dispersión")

ax3.plot(nombres,valores)
ax3.set_title("Gráfica de líneas")

tamanio= np.array(valores)/sum(valores)

ax4.pie(tamanio, labels= nombres, autopct='%.2f%%', startangle=45)

plt.show()
```
